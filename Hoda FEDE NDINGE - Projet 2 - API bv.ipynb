```python

import requests
def get_pokemon_stats():
    # Demander le nom ou l'ID du Pokémon
    pokemon_name_or_id = input("Entrez le nom ou l'ID d'un Pokémon : ").lower()

     # URL pour obtenir les données du Pokémon depuis PokeAPI
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name_or_id}"

    try:
        # Requête GET pour récupérer les informations des Poke
        response = requests.get(url)
        response.raise_for_status()  # Vérifie si la requête est réussie

        # JSON
        data = response.json()

        # Pour avoir les statistiques
        name = data['name'].capitalize()
        stats = {stat['stat']['name']: stat['base_stat'] for stat in data['stats']}

        # Afficher les statistiques
        print(f"\nStatistiques de {name} :")
        print(f"Points de vie (HP) : {stats.get('hp', 'Non disponible')}")
        print(f"Attaque : {stats.get('attack', 'Non disponible')}")
        print(f"Défense : {stats.get('defense', 'Non disponible')}")
        print(f"Attaque spéciale : {stats.get('special-attack', 'Non disponible')}")
        print(f"Défense spéciale : {stats.get('special-defense', 'Non disponible')}")
        print(f"Vitesse : {stats.get('speed', 'Non disponible')}")
       
        #Afficher resultats
    except requests.exceptions.HTTPError:
        print("Le Pokémon que vous avez entré est introuvable. Veuillez vérifier le nom ou l'ID.")
    except requests.exceptions.RequestException as e:
        print(f"Une erreur s'est produite : {e}")

# Execute le code
get_pokemon_stats()

#verification
```

```python
#https://pokeapi.co/api/v2/type/{type} : Récupère les infos sur un type de Pokémon
#Écrivez une fonction qui demande au user : Le type des Pokémon (fire, dragon, etc.) et calcule le nombre des Pokémon de ce type fire, la moyenne des points de vie (HP)
```

```python
import requests
 
def get_pokemon_by_type():

    type_name = input("Entrez le type de Pokémon en anglais : ")

    url = f'https://pokeapi.co/api/v2/type/{type_name}'

    response = requests.get(url)

    if response.status_code == 200:

        data = response.json()

        pokemon_count = len(data['pokemon'])

        total_hp = 0

        for p in data['pokemon']:

            pokemon_url = p['pokemon']['url']

            pokemon_response = requests.get(pokemon_url)

            if pokemon_response.status_code == 200:

                pokemon_data = pokemon_response.json()

                total_hp += pokemon_data['stats'][0]['base_stat']

        average_hp = total_hp / pokemon_count if pokemon_count > 0 else 0

        #Resultats
        print(f"Le Type/Genre: {type_name}")

        print(f"Nombre de Pokémon: {pokemon_count}")

        print(f"Moyenne HP: {average_hp:.2f}")
 
    else:

        print("Erreur : Type non trouvé")
#Execute le code        

get_pokemon_by_type()
```

```python
###### Réponse Q2 : Proposer une fonction qui permet de gérer des millions de requêtes de manière efficace et simulez à l’aide d’un script une charge très élevée de call vers plusieurs rootes : ["https://api.example.com/data1", "https://api.example.com/data2", "https://api.example.com/data3"] ?
#Indice : concurrence et parallélisme en Python, asyncio 
```

```python
import aiohttp
import asyncio
import time

# Fonction asyn pour recup les donnees 
async def fetch(session, url):
    async with session.get(url) as response:
        return await response.text()

# Gerer les requetes
async def fetch_all(urls, max_connections=100):
    # Sémaphore--> limite le nmbre de connex
    semaphore = asyncio.Semaphore(max_connections)

    async with aiohttp.ClientSession() as session:
        tasks = []
        for url in urls:
            async with semaphore:
                tasks.append(fetch(session, url))
        
        responses = await asyncio.gather(*tasks)
        return responses

async def main():
    urls = ["https://api.example.com/data1", "https://api.example.com/data2", "https://api.example.com/data3"] * 1000000

    # Temps que met les requetes 
    start_time = time.time()

    # Syncro toutes les données 
    results = await fetch_all(urls, max_connections=500)

    print(f"Récupéré {len(results)} réponses en {time.time() - start_time} secondes")

# Execute
if __name__ == "__main__":
    asyncio.run(main())

# Cette ligne de code, doit etre la dans le block n°1, pour pouvoir avoir cette fonction dans l'exercice
```

```python
import requests

# Fonction pour obtenir les statistiques d'un Pokémon via l'API PokeAPI
def get_pokemon_stats(pokemon_name_or_id):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name_or_id.lower()}"
    
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        name = data['name'].capitalize()
        stats = {stat['stat']['name']: stat['base_stat'] for stat in data['stats']}
        
        # Retourner les statistiques nécessaires pour le combat
        return {'name': name, 'attack': stats.get('attack', 0), 'defense': stats.get('defense', 0)}
    
    except requests.exceptions.HTTPError:
        print(f"Le Pokémon {pokemon_name_or_id} est introuvable.")
        return None
    except requests.exceptions.RequestException as e:
        print(f"Une erreur s'est produite : {e}")
        return None

# Fonction pour simuler le combat entre deux Pokémon sur 5 tours
def battle_pokemons():
    # Mettre le nom des PK 
    pokemon1_name = input("Entrez le nom du premier Pokémon: ")
    pokemon2_name = input("Entrez le nom du deuxième Pokémon: ")
    
    # Récup les stats dans l'API
    stats1 = get_pokemon_stats(pokemon1_name)
    stats2 = get_pokemon_stats(pokemon2_name)
    
    # Voir si les 2 sont dans la BD
    if stats1 is None or stats2 is None:
        print("Un ou les deux Pokémons n'ont pas été trouvés.")
        return
    
    # Initialiser les scores de dégâts pour les 2 PK
    pokemon1_damage = 0
    pokemon2_damage = 0
    turns = 5

    # Simuler les 5 tours de combat
    for turn in range(1, turns + 1):
        print(f"\n--- Tour {turn} ---")

        # Dommages infligés PK 1 à PK 2
        damage1 = max(1, stats1['attack'] - stats2['defense'])  # Les dégâts doivent être au moins de 1
        pokemon1_damage += damage1
        print(f"{stats1['name']} inflige {damage1} dommage à {stats2['name']}.")

        # Dommages infligés par PK 2 à PK 1
        damage2 = max(1, stats2['attack'] - stats1['defense'])
        pokemon2_damage += damage2
        print(f"{stats2['name']} inflige {damage2} dommage à {stats1['name']}.")

    # Le vainqueur apres 5 T
    print("\n--- Résultats finaux ---")
    print(f"{stats1['name']} a infligé un total de {pokemon1_damage} dommage.")
    print(f"{stats2['name']} a infligé un total de {pokemon2_damage} dommage.")
    
    if pokemon1_damage > pokemon2_damage:
        print(f"{stats1['name']} remporte le combat !")
    elif pokemon2_damage > pokemon1_damage:
        print(f"{stats2['name']} a remporter le combat !")
    else:
        print("Le combat se termine par une égalité !")

# Exécuter
battle_pokemons()
```
